Naukri-Data Science- Bangalore

```python
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
driver=webdriver.Chrome(r"C:\Users\Anurag\Downloads\chromedriver.exe")
```

```python
url= 'https://www.naukri.com/'
driver.get(url)
```

```python
search_job= driver.find_element_by_id('qsb-keyword-sugg')
```

```python
search_job.send_keys("Data Scientist")
```

```python
search_loc= driver.find_element_by_id('qsb-location-sugg')
```

```python
search_loc.send_keys("Bangalore")
```

```python
search_btn=driver.find_element_by_xpath("//button[@class='btn']")
search_btn.click()
```

```python
urls=[]
for i in driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']"):
    urls.append(i.get_attribute("href"))
```

```python
urls
```

```python
job_titles=[]
job_location=[]
company_name=[]
job_description=[]
```

```python
for i in urls[:10]:
    driver.get(i)
    time.sleep(3)
    
    try:
        job_name=driver.find_element_by_xpath("//h1[@class='jd-header-title']")
        job_titles.append(job_name.text)
    except NoSuchElementException:
        job_titles.append('-')
        
    try:
        location=driver.find_element_by_xpath("//i[@class='location expand']")
        job_location.append(location.text)
    except NoSuchElementException:
        job_location.append('-')
        
    try:
        company=driver.find_element_by_xpath("//div[@class='jd-header-comp-name']")
        company_name.append(company.text)
    except NoSuchElementException:
        company_name.append('-')
        
    try:
        job_desc=driver.find_element_by_xpath("//section[@class='job-desc']")
        job_description.append(job_desc.text)
    except NoSuchElementException:
        job_description.append('-')
        
```

```python
print(len(job_titles),len(job_location),len(company_name),len(job_description))
```

```python
naukri=pd.DataFrame({})
naukri['Job_Titles']=job_titles
naukri['Job_Location']=job_location
naukri['Name']=company_name
naukri['Description']=job_description
```

```python
naukri
```
