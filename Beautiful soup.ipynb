book review

```python
page=requests.get("https://bookpage.com/reviews")
page
```

```python
page.content
```

```python
bsp=BeautifulSoup(page.content)
bsp
```

```python
print(bsp.prettify())
```

```python
name=bsp.find_all('h4',class_="italic")
name
```

```python
title=[]
for i in name:
    title.append(i.text.replace('\n', ' '))
title
```

```python
gen=bsp.find_all('p',class_="genre-links hidden-phone")
gen
```

```python
genre=[]
for i in gen:
    genre.append(i.text.replace('\n', ' '))
genre
```

```python
rev=bsp.find_all('p',class_="excerpt")
rev
```

```python
review=[]
for i in rev:
    review.append(i.text.replace('\n', ' '))
review
```

```python
print(len(title),len(genre),len(review))
```

```python
import pandas as pd
books=pd.DataFrame({})
books['Title']=title
books['Genre']=genre
books['Review']=review
```

```python
books
```

Sans Francisco weather

```python
weather=requests.get("https://forecast.weather.gov/MapClick.php?lat=37.7771&lon=-122.4196#.YJz1wbUzZPY")
weather
```

```python
weather.content
```

```python
bsoup=BeautifulSoup(weather.content)
bsoup
```

```python
time=bsoup.find_all('div',class_="col-sm-2 forecast-label")
time
```

```python
duration=[]
for i in time:
    duration.append(i.text.replace('\n', ' '))
duration
```

```python
forecast=bsoup.find_all('div',class_="col-sm-10 forecast-text")
forecast
```

```python
forecasting=[]
for i in forecast:
    forecasting.append(i.text.replace('\n', ' '))
forecasting
```

```python
print(len(duration),len(forecasting))
```

```python
import pandas as pd
Francisco=pd.DataFrame({})
Francisco['Day']=duration
Francisco['Forecast']=forecasting
```
