```python
def bar_plot(x,y):
    df1_plot=df.groupby([x,y]).size().reset_index().pivot(columns=y,index=x,values=0)
    df1_plot=df1_plot.fillna(0)
    plt.figure(figsize=(10,10))                #Bivariate Analysis
    df1_plot.plot(kind='bar')
    plt.show()
```

```python
def applications(x):
    list6=[]
    for i in df[x]:
        i.split(',')
        list6.append(i)
        
    res = [sub.split() for sub in list6]
    flat_list6=[]                                   
    for sublist in res:
        for item in sublist:
            flat_list6.append(item)
            
    flat_list6=[re.sub(",","",i) for i in flat_list6]
            
    
    ax=sns.countplot(flat_list6)
    plt.xticks(rotation=90)
    ax.set(xlabel=x, ylabel = "Count")
    plt.show()
```

```python
applications('Security of customer financial information')
```

```python
applications('Quickness to complete purchase')
```

```python
applications('Speedy order delivery')
```

```python
for i in ['Change in website/Application design',
'Frequent disruption when moving from one page to another',
'Website is as efficient as before']:
    applications(i)
```

```python
applications('Which of the Indian online retailer would you recommend to a friend?')
```

a)When comes to website design Amazon and then Flipkart is performing much better compare to rest of others.

b)During moving one page to other frequent disruptions occurs and Paytm is performing better.
