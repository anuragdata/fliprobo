```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(3)

# Opening the homepage of Amazon.in
url = "https://www.amazon.in/New-Apple-iPhone-11-64GB/product-reviews/B08L8DV7BX/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
driver.get(url)
```

```python
review_url=[]
while True:
    try:
        for j in driver.find_elements_by_xpath("//a[@class='a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold']"):
            review_url.append(j.get_attribute('href'))
        driver.find_element_by_xpath('//ul[@class="a-pagination"]/li[2]/a').click()
        time.sleep(3)
        print('reviews url found')
    except:
        print('no reviews')
        break
```

```python
len(review_url)
```

```python
reviews=[]     # full reviews
stars=[]
for i in review_url:
    driver.get(i)
    time.sleep(3)
    a=driver.find_element_by_xpath('//span[@class="a-size-base review-text review-text-content"]')
    reviews.append(a.text)
    b=driver.find_element_by_xpath('//i[@data-hook="review-star-rating"]')
    stars.append(b.get_attribute('class'))
```

```python
df=pd.DataFrame({'Reviews':reviews,'Ratings':stars})
```

```python
df
```

```python
df.to_csv('reviews.csv')
```

```python
import re
import nltk

from nltk import word_tokenize, pos_tag
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import sent_tokenize
from nltk.corpus import wordnet
```

```python
#create Lemmatizer object
lemma = WordNetLemmatizer()

def lemmatize_word(tagged_token):
    """ Returns lemmatized word given its tag"""
    root = []
    for token in tagged_token:
        tag = token[1][0]
        word = token[0]
        if tag.startswith('J'):
            root.append(lemma.lemmatize(word, wordnet.ADJ))
        elif tag.startswith('V'):
            root.append(lemma.lemmatize(word, wordnet.VERB))
        elif tag.startswith('N'):
            root.append(lemma.lemmatize(word, wordnet.NOUN))
        elif tag.startswith('R'):
            root.append(lemma.lemmatize(word, wordnet.ADV))
        else:          
            root.append(word)
    return root

def lemmatize_doc(document):
    """ Tags words then returns sentence with lemmatized words"""
    lemmatized_list = []
    tokenized_sent = sent_tokenize(document)
    for sentence in tokenized_sent:
        no_punctuation = re.sub(r"[`'\",.!?()]", " ", sentence)
        tokenized_word = word_tokenize(no_punctuation)
        tagged_token = pos_tag(tokenized_word)
        lemmatized = lemmatize_word(tagged_token)
        lemmatized_list.extend(lemmatized)
    return " ".join(lemmatized_list)

#apply our functions
df["preprocessed"] = df["preprocessed"].apply(lambda row: lemmatize_doc(row))
```
