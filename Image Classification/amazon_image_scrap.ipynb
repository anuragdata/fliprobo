```python
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

#Importing requests
import requests

# importing regex
import re
```

```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(3)

# Opening the homepage of Amazon.in
url = "https://www.amazon.in/s?k=jeans&ref=nb_sb_noss_2"
driver.get(url)
```

```python
image_url=[]
while True:
    try:
        for j in driver.find_elements_by_xpath("//img[@class='s-image']"):
            image_url.append(j.get_attribute('src'))
        time.sleep(3)
        print('image url found')
    except:
        print('no image')
        break
```

```python
len(image_url)
```

```python
imgg=image_url[0:300]
```

```python
imgg
```

```python
            
for i in range(len(imgg)):
    if i >= 275:
        break
    print("Downloading {0} of {1} images" .format(i, 275))
    response= requests.get(imgg[i])
    file = open(r"C:\Users\Anurag\jeans"+str(i)+".jpg", "wb")
    file.write(response.content)
```
